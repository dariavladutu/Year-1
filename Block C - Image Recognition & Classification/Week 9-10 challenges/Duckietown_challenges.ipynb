## Drive Straight (Code duckietown to move in a straight line.)


```python
#Step 1: Setting up the environment 
import gym
import gym_duckietown
from IPython.display import display, clear_output
from PIL import Image

env = gym.make('Duckietown-small_loop-v0')
env.reset()

# Step 2: Displaying an image in Jupyter notebook
def show_array(arr):
    img = Image.fromarray(arr, 'RGB')
    img = img.resize((320, 240))
    display(img)

# Step 3: Running the simulation
done = False
while not done:
    # Selecting a random action
    action = env.action_space.sample()
    # Performing the action, advancing the simulation by one step, and retrieving the observation (state), reward (result), done, and info
    observation, reward, done, info = env.step(action)
    
    # Renderring the frame to an array of RGB values (visualization of simulated environment)
    frame = env.render(mode='rgb_array')

    # Clearing the previous output and updating the displayed image
    clear_output(wait=True)
    show_array(frame)
    
    # Resetting the environment to its initial state if done
    if done:
        env.reset()

# Step 4: Closing the environment after simulation is done
env.close()
```

Complete Simple Track (Navigate a basic track with four 90-degree turns. Can your Duckiebot stay within the lines?)

```python
# Step 1: Setting up the environment
import time
import gym
import gym_duckietown
from IPython.display import display, clear_output
from PIL import Image
import numpy as np

env = gym.make('Duckietown-small_loop_cw-v0')
env.reset()

# Step 2: Displaying an image in Jupyter notebook
def show_array(arr):
    img = Image.fromarray(arr, 'RGB')
    img = img.resize((320, 240))
    display(img)

# Step 3: Running the simulation
# Function to perform a straight move for a specified duration
def move_straight(duration=4.0):
    action = np.array([0.3, 0.0])  # Move straight
    start_time = time.time()
    while time.time() - start_time < duration:
        observation, reward, done, info = env.step(action)
        clear_output(wait=True)
        show_array(observation)

# Function to perform a 90-degree turn
def turn_90_degrees():
    action = np.array([0.3, np.pi / 2])  # Turn 90 degrees
    env.step(action)

# Navigate the basic track with four 90-degree turns
for i in range(4):
    # Move straight for a short duration
    move_straight()
    
    # Perform a 90-degree turn
    turn_90_degrees()

# Reset the environment
env.reset()

env.close()
```

Race Simple Track 
(It's time for speed. Who will top the leaderboard with the fastest lap?)

```python
import time
import gym
import gym_duckietown
import numpy as np

# Create the environment
env = gym.make('Duckietown-small_loop_cw-v0')
env.reset()

# Function to race on the track with maximum speed 
def race_simple_track(duration=60.0):
    action = np.array([1.0, 0.0])  # Move at maximum speed
    start_time = time.time()
    return env, action, duration, start_time
    
def run_race(env, action, duration, start_time):
    try:
        while time.time() - start_time < duration:
            observation, _, done, _ = env.step(action)
            
            if done:
                break
            
            # Render the environment periodically to reduce rendering overhead
            if int(time.time() - start_time) % 2 == 0:
                env.render()

    finally:
        env.close()

def race_track(duration=60.0):
    env, action, duration, start_time = race_simple_track(duration)
    run_race(env, action, duration, start_time)

# Race on the simple track for 60 seconds (adjust as needed)
race_track()
```

Lane Detection (Implement lane detection algorithms to keep your Duckiebot on track.)

```python
import time
import gym
import gym_duckietown
from IPython.display import display, clear_output
from PIL import Image

# Create the environment
env = gym.make('Duckietown-loop_empty-v0')
env.reset()

def show_array(arr):
    img = Image.fromarray(arr, 'RGB')
    img = img.resize((320, 240))
    display(img)

# Function to follow the lane using the built-in lane following controller
def follow_lane():
    while True:
        action = [0.3, 0.0]  # Move forward
        observation, reward, done, info = env.step(action)
        # Display
        clear_output(wait=True)
        show_array(observation)

        if done:
            break

# Perform lane following
follow_lane()

# Reset the environment
env.reset()

env.close()
```
