## Demographic parity

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for visualization
np.random.seed(0)
num_samples = 1000

# Equal Selection Parity: Create synthetic prediction scores for two groups
prediction_scores_A = np.random.normal(loc=0.6, scale=0.1, size=num_samples)
prediction_scores_B = np.random.normal(loc=0.6, scale=0.1, size=num_samples)

# Demographic Parity: Create synthetic approved loan distributions for two groups
approved_loans_A = np.random.binomial(n=1, p=0.8, size=num_samples)
approved_loans_B = np.random.binomial(n=1, p=0.8, size=num_samples)

# Visualize Equal Selection Parity
plt.figure(figsize=(8, 6))
plt.hist(prediction_scores_A, bins=30, color='blue', alpha=0.5, label='Group A', density=True)
plt.hist(prediction_scores_B, bins=30, color='orange', alpha=0.5, label='Group B', density=True)
plt.axvline(x=0.5, color='r', linestyle='--', label='Threshold')
plt.xlabel('Prediction Score')
plt.ylabel('Density')
plt.title('Equal Selection Parity')
plt.legend()
plt.grid(True)
plt.show()

# Visualize Demographic Parity
plt.figure(figsize=(8, 6))
plt.hist(approved_loans_A, bins=2, color='blue', alpha=0.5, label='Group A', density=True)
plt.hist(approved_loans_B, bins=2, color='orange', alpha=0.5, label='Group B', density=True)
plt.xlabel('Loan Approval (0=Not Approved, 1=Approved)')
plt.ylabel('Density')
plt.title('Demographic Parity')
plt.legend()
plt.grid(True)
plt.xticks([0, 1])
plt.show()
```

## Equalized opportunities

## Predictive Parity

```python
# Generate synthetic data for visualization
np.random.seed(0)
num_samples = 1000

# Generate synthetic binary classification outcomes for two groups
binary_outcomes_A = np.random.binomial(n=1, p=0.8, size=num_samples)  # Binary outcomes for group A
binary_outcomes_B = np.random.binomial(n=1, p=0.8, size=num_samples)  # Binary outcomes for group B

# Calculate positive predictive value (PPV) for each group
def calculate_ppv(binary_outcomes):
    true_positives = np.sum(binary_outcomes == 1)
    if len(binary_outcomes) == 0:
        return 0  # Handle division by zero
    return true_positives / len(binary_outcomes)

ppv_group_A = calculate_ppv(binary_outcomes_A)  # PPV for group A
ppv_group_B = calculate_ppv(binary_outcomes_B)  # PPV for group B

# Visualize Predictive Parity
plt.figure(figsize=(8, 6))
plt.bar(['Group A', 'Group B'], [ppv_group_A, ppv_group_B], color=['blue', 'orange'], alpha=0.7)
plt.ylim(0, 1)
plt.ylabel('Positive Predictive Value (PPV)')
plt.title('Predictive Parity')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
