### Matplotlib hands-on

The dataset was downloaded from here: 
[https://data.ontario.ca/dataset/covid-19-vaccine-data-in-ontario/resource/c08620e0-a055-4d35-8cec-875a459642c3]

```python
import matplotlib.pyplot as plt
import pandas as pd
vac_cases= pd.read_csv('cases_by_age_vac_status.csv', parse_dates = ['date'], index_col = ['date'])
```

```python
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
fig.set_size_inches([10, 8])
first_yr_of_data = vac_cases['2021-01-01':'2022-10-24']
ax.plot(first_yr_of_data.index, vac_cases['cases_unvac_rate_per100K'], color='b', marker = 'o', linestyle = '--')
ax.set_xlabel('Time')
ax.set_ylabel('Cases of unvaccinated per 100k pop')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(first_yr_of_data.index, vac_cases['cases_unvac_rate_per100K'], color='b')
ax2 = ax.twinx()
ax2.plot(first_yr_of_data.index, vac_cases['cases_full_vac_rate_per100K'], color='r')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(vac_cases.index, vac_cases['cases_unvac_rate_per100K'])
ax.set_ylabel('Cases of unvaccinated per 100k pop')
ax.set_xlabel('2022 - 2023')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(vac_cases['cases_unvac_rate_per100K'], label = 'Unvac rate', color='pink', bins = 5)
ax.hist(vac_cases['cases_full_vac_rate_per100K'], label = 'Vac rate',  bins = 5)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(vac_cases['cases_full_vac_rate_per100K'], vac_cases['cases_partial_vac_rate_per100K'])
ax.set_xlabel('Full vac rate per 100k pop')
ax.set_ylabel('Partial vac rate per 100k pop')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot([vac_cases['cases_full_vac_rate_per100K']])
ax.set_xticklabels(["Full vac rate"])
ax.set_ylabel("Per 100k")
plt.show()
```
