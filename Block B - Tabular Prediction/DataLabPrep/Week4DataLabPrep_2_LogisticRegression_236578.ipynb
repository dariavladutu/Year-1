### Fine-tuning the model (DataCamp)

```python
# Confustion matrix
'''                         Predicted:            Predicted:
                            Positive              Negative

Actual: Positive           True Positive(TP)      False Negative(FN)
Actual: Negative           False Positive(FP)     True Negative(TN)'''
```

Precision:
TRUE POSITIVES/ (TRUE POSITIVES + FALSE POSITIVES)
- High precision = lower false positive rates

Recall:
TRUE POSITIVES/ (TRUE POSITIVES + FALSE NEGATIVES)
- High recall = lower false negative rate

F1 score:
2*(precision * recall)/ precision + recall
- High F1 score = high F1 score, high precision and recall (∈ [0,1])

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_score, f1_score
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
NAC_data =  pd.read_csv('NAC_data.csv')
NAC_df = pd.DataFrame(NAC_data)
NAC_df.head()
```

```python
# Extracting features and target
X = NAC_df['Age'].values
y = NAC_df['Market value'].values
X = X.reshape(-1, 1)
# Impute NaN values with the mean
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
# Installing K Nearest Neighbor Classifier model
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_imputed,y)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.4, random_state=42)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
# Generating the confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
```

```python
# Calculated f1-score and precision sepparately as the function encountered multiple 'zero-division' instances
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)
report = classification_report(y_test, y_pred)
print("Precision:", precision)
print("F1-Score:", f1)
print("Classification Report:")
print(report)
```

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Instantiate the model
logreg = LogisticRegression()

# Fit the model
logreg.fit(X_train, y_train)

# Predict probabilities
y_pred_probs = logreg.predict_proba(X_test)[:, 1]

print(y_pred_probs[:10])
```

Split the dataset into features (X) and target variable (y)

Split the data into training and testing sets.

```python
X_bc_train, X_bc_test, y_bc_train, y_bc_test = train_test_split(X_bc, y_bc, test_size=0.2, random_state=42)

# Display the missing values and the shape of the dataset
missing_values, X_bc_train.shape, X_bc_test.shape, y_bc_train.shape, y_bc_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Train the model on the training set

```python
# Instantiate the model
bc_logreg = LogisticRegression(max_iter=10000)

# Fit the model
bc_logreg.fit(X_bc_train, y_bc_train)
```
